```python
total_cuenta = float(input("Ingrese el total de la cuenta:"))
p_propina = float(input("Ingrese el % de propina que desea dejar:"))
propina = total_cuenta * (p_propina / 100)
total_a_pagar = total_cuenta + propina
print(f"La propina es: {propina}")
print(f"El total a pagar es: {total_a_pagar}")
```

```python
n_personas = int(input("Ingrese el número de personas:"))
if n_personas > 0:
    total_por_persona = total_a_pagar / n_personas
    print(f"El total a pagar por persona es: {total_por_persona}")
else:
    print("El número de personas debe ser mayor que cero.")
```
