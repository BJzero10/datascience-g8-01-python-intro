#### Funciones u operadores en strings

- Longitud de mi cadena (len)
- Convertir a mayuscula (upper)
- Convertir a minuscula (lower)
- Reemplazar un valor de la cadena (replace)
- "texto      "  -> (strip) -> "texto"
- "texto1, texto2, texto3" -> (split, ",") -> [texto1, texto2 , texto3]

```python
nombre = "Jesus"
print("Cuantas letras tiene mi nombre: ", len(nombre))
print(len("peru"))
```

```python
print(nombre.upper())
```

```python
print(nombre.lower())
```

```python
# replace acepta dos parametros, primer parametro: lo que quiero reemplazar, segundo parametro: el valor por el que voy a reemplazar   
print(
    nombre.replace("us", "")
)
```

```python
nombre = "Jesus                   "
print(len(nombre)) # 26 caracteres

nombre_limpio = nombre.strip() # limpiando resulta en 7 caracteres

print(len(nombre_limpio))

nombre = "              JESUS            " # 33 caracteres
print(len(nombre))

nombre_limpio = nombre.strip() # 7 caracteres
print(len(nombre_limpio))
```

```python
nombres = "Jesus Sullon"

nombres_separados = nombres.split(" ")
print(nombres_separados)

nombres = "JesusxxSullon"
nombres_separados = nombres.split("xx")
print(nombres_separados)
```
